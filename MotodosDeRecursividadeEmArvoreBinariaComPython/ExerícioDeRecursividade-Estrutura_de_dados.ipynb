```python
#Aluno: Ruan Carlos ****
#Matricula: 0133****
#disciplina: Estrutura de dados


# Utilizando a classe Node para representar os nós da árvore
class Node:
    def __init__(self, key):
        # Cada nó possui uma chave, uma referência
        # ao nó à esquerda e outra à direita
        self.key = key
        self.left = None # Nó da direita
        self.right = None # Nó da esquerda

# Definindo a classe BinaryTree para representar a árvore binária
class BinaryTree:
    def __init__(self):
        # Inicializando a árvore com a raiz nula
        self.root = None #Utilizando o objeto None para atribuir valor Inicial
                         #nulo na árvore
    def insert(self, key):
        # Método público para inserir uma chave na árvore
        self.root = self._insert(self.root, key)

#O método _insert é um método privado que realiza a inserção
#de um novo nó na árvore.
    def _insert(self, root, key):

        if root is None:
            return Node(key)
            #Se o nó atual for None,
            #cria um novo nó com a chave fornecida.



#Se a chave for maior, chama recursivamente _insert para o nó à direita.
#Retorna o nó atual.
        if key < root.key:
            root.left = self._insert(root.left, key)
            #Se a chave for menor que a chave do nó atual,
            #chama recursivamente _insert para o nó à esquerda.
        elif key > root.key:
            root.right = self._insert(root.right, key)
            #Se a chave for maior, chama recursivamente _insert
            #para o nó à direita.

        return root

#O método remove é utilizado para remover um nó da árvore,
#chamando o método privado _remove.
    def remove(self, key):
        # Método público para remover uma chave da árvore
        self.root = self._remove(self.root, key)


    def _remove(self, root, key):
        #O método remove é utilizado para remover um nó da
        #árvore, chamando o método privado _remove.
        if root is None:
            return root

        if key < root.key:
            root.left = self._remove(root.left, key)
        elif key > root.key:
            root.right = self._remove(root.right, key)
        else:
            # Caso 1: Nó com apenas um filho ou nenhum filho
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
#Se o nó a ser removido possui apenas um filho ou nenhum filho, o próprio nó é
#removido e substituído pelo seu único filho (se existir). Isso é feito
#atribuindo root.right a root se root.left for None, e vice-versa.

            # Caso 2: Nó com dois filhos
            root.key = self._min_value_node(root.right).key
            root.right = self._remove(root.right, root.key)
#Se o nó a ser removido possui dois filhos, o valor do nó é substituído pelo
#menor valor encontrado na subárvore à direita (o sucessor in-order).
#Em seguida, o nó com esse menor valor é removido, aplicando recursivamente
#a função _remove na subárvore à direita.

        return root

    #O método _min_value_node encontra e retorna o nó com o menor
    #valor em uma subárvore.
    def _min_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

#O método inorder_traversal realiza uma travessia em ordem na árvore,
#imprimindo os valores dos nós.
    def inorder_traversal(self, root):
        if root:
            self.inorder_traversal(root.left)
            print(root.key, end=" ")
            self.inorder_traversal(root.right)

#O programa principal cria uma instância da classe BinaryTree.
#Um loop while permite que o usuário interaja com a árvore, escolhendo
#entre inserir, remover, listar ou sair.
if __name__ == "__main__":
    tree = BinaryTree()

    while True:
        # Menu de interação com o usuário
        print("\n## ÁRVORE BINÁRIA ##")
        print("1 - Inserir")
        print("2 - Remover")
        print("3 - Listar árvore")
        print("4 - Sair")

        # Solicitação da escolha do usuário
        choice = int(input("Informe a opção desejada: "))

        if choice == 1:
            # Opção de inserção
            key = int(input("Informe a chave para inserção: "))
            tree.insert(key)
        elif choice == 2:
            # Opção de remoção
            key = int(input("Informe a chave para remoção: "))
            tree.remove(key)
        elif choice == 3:
            # Opção de listagem (inorder traversal)
            print("Listar árvore: ", end="")
            tree.inorder_traversal(tree.root)
            print()
        elif choice == 4:
            # Opção de sair do programa
            print("Sair")
            break
        else:
            # Opção inválida
            print("Opção inválida!")
```
